## Extracting Features from Images
We have three ways of extracting features from images.
1) GrayScale Pixel values as Features
2) Mean Pixel Values of RGB channels
3) Edge Feature Extraction

### 2) Mean Pixel Values of RGB channels

```python
image1 = imread(r"D:\PYTHON_Programmes\Python_Projects\2105800_ML_Projects\Image_Classification_Projects\catsAndDogs40\train\cat\1.jpg", as_gray=False)
new_feature_matrix = np.zeros((160,160))
for i in range(0,image1.shape[0]):
    for j in range(0,image1.shape[1]):
        new_feature_matrix[i][j] = ((int(image1[i,j,0])+int(image1[i,j,1])+int(image1[i,j,2]))/3)
image1_features = np.reshape(new_feature_matrix, (160,160))
image1_features.shape
```

### Now, we can try to understand the various ways in which we can extract the images from DR Samples such that all the trends and features are clearly visible.

```python
retina_image1 = imread(r"D:\PYTHON_Programmes\Python_Projects\2105800_ML_Projects\Image_Classification_Projects\DR_Dataset\images.jpeg", as_gray=True)
retina_image2 = imread(r"D:\PYTHON_Programmes\Python_Projects\2105800_ML_Projects\Image_Classification_Projects\DR_Dataset\images.jpeg", as_gray=False)
imshow(retina_image1)
```

```python
imshow(retina_image2)
retina_image2.shape
```

```python
retina_edge_h = prewitt_h(retina_image1)
retina_edge_v = prewitt_v(retina_image1)

fig, (ax1, ax2)=plt.subplots(1, 2, figsize=(13,7))

ax2.imshow(retina_edge_v)
ax2.axis("on")
ax2.set_title("Retina_Exudate_Vertical_Edge_Feature")

ax1.imshow(retina_edge_h)
ax1.axis("on")
ax1.set_title("Retina_Exudate_Horizonal_Edge_Feature")

fig.savefig(r"D:\PYTHON_Programmes\Python_Projects\2105800_ML_Projects\Image_Classification_Projects\DR_Dataset\images_edge_features_extracted1.jpg")
```

From the above analysis, we can conclude that the leakage of Exudates from the newly created blood-vessels can be easily identified and is clearly visible in the above picture where the Edge Color of the exudates and the color of the retina are different. This allows us to be able to clearly identify the start of Diabetic Retinopathy. There may be other methods mentioned above which may be used to identify different conditions.

Therefore, We cannot detect Haemorrage with the help of Edge Feature Extraction. Here, we may use the mean pixel values of RBG for distinguishing the **Red pigmentation of the Vitreous Humor** inside the retina and then understand the trend and make predictions or classifications about the same. 

Other techniques like the HOG (Histogram of Oriented Gradients) can be used to detect the excess growth/newly growing blood vessels to point out the DR condition at an early stage. For Non-Proliferative DR, this is possible, but in advanced stages of DR, i.e., Proliferative DR, it becomes difficult to save vision of the patient.
